```python
theta = np.linspace(0, np.pi, 360)
y = np.sin(theta)
x = np.cos(theta)
plt.plot(x, y)
plt.plot(theta, np.cos(theta))
plt.xlabel('theta')
```

```python
x = np.array([1, 2, 3, 4])
print("3^x =", np.power(3., x))
```

```python
print("log10(x) =", np.log10(x))
```

# Comparisons, Masks, and Boolean Logic

```python
x = np.array([1, 2, 3, 4, 5])
x < 3  # less than
```

```python
seed = 1701 #change the seed to get different random numbers
```

```python
rng = np.random.default_rng(seed=seed) # seed is fix if we want to get the same random numbers every time we run the code
x = rng.integers(10, size=(3, 4))
x, x < 6
```

```python
np.all(x < 8, axis=1) # check if all elements in each row(axis=1) are less than 8
```

```python
A = np.array([1, 0, 1, 0, 1, 0], dtype=bool)
B = np.array([1, 1, 1, 0, 1, 1], dtype=bool)
A | B
```

```python
A & B
```

```python
np.logical_xor(A, B), np.logical_not(A), np.logical_or(A, B), np.logical_and(A, B)
```

```python
bool(42 and 0), bool(42 or 0), bool(42 and 42)
```

```python
bin(59), bin(59)[2:], int('111011', 2)
```

```python
bin(42 & 59), bin(42 | 59), bin(42 ^ 59)
```

```python
A = np.array([1, 0, 1, 0, 1, 0], dtype=bool)
B = np.array([1, 1, 1, 0, 1, 1], dtype=bool)
A | B
```
